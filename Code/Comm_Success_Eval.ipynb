```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
result_base_dir = '/media/jmorisse/Elements/API_Project/Results'
seeds = [1,2,3,4,5]
```

```python
text_base_dir = result_base_dir+'/Text'
train_accs = []
train_losses = []
val_accs = []
val_losses = []
test_accs = []
test_losses = []
for seed in seeds:
    seed_dir = text_base_dir+'/Seed_'+str(seed)
    
    with open(seed_dir+'/train_accs.txt', 'r') as f:
        for line in f:
            seed_accs = [float(acc) for acc in line[1:-2].split(',')]
            train_accs.append(seed_accs)
    
    with open(seed_dir+'/train_losses.txt', 'r') as f:
        for line in f:
            seed_losses = [float(loss) for loss in line[1:-2].split(',')]
            train_losses.append(seed_losses)
    
    with open(seed_dir+'/val_accs.txt', 'r') as f:
        for line in f:
            seed_accs = [float(acc) for acc in line[1:-2].split(',')]
            val_accs.append(seed_accs)
    
    with open(seed_dir+'/val_losses.txt', 'r') as f:
        for line in f:
            seed_losses = [float(loss) for loss in line[1:-2].split(',')]
            val_losses.append(seed_losses)
    
    with open(seed_dir+'/test_acc.txt', 'r') as f:
        for line in f:
            seed_accs = float(line)
            test_accs.append(seed_accs)
    
    with open(seed_dir+'/test_loss.txt', 'r') as f:
        for line in f:
            seed_losses = float(line)
            test_losses.append(seed_losses)
```

```python
train_accs_mean = np.mean(np.array(train_accs), axis=0)
train_accs_std = np.std(np.array(train_accs), axis=0)
train_losses_mean = np.mean(np.array(train_losses), axis=0)
train_losses_std = np.std(np.array(train_losses), axis=0)
val_accs_mean = np.mean(np.array(val_accs), axis=0)
val_accs_std = np.std(np.array(val_accs), axis=0)
val_losses_mean = np.mean(np.array(val_losses), axis=0)
val_losses_std = np.std(np.array(val_losses), axis=0)
test_accs_mean = np.mean(np.array(test_accs), axis=0)
test_accs_std = np.std(np.array(test_accs), axis=0)
test_losses_mean = np.mean(np.array(test_losses), axis=0)
test_losses_std = np.std(np.array(test_losses), axis=0)
```

```python
print('Mean test acc:', test_accs_mean)
print('Std test acc:', test_accs_std)
print('Mean test loss:', test_losses_mean)
print('Std test loss:', test_losses_std)

total_nr_epochs = 200
plt.figure(figsize = (15,7))
line1, = plt.plot(range(0,201),train_losses_mean)
plt.fill_between(x=range(0,201), y1=train_losses_mean-train_losses_std, y2=train_losses_mean+train_losses_std, alpha=0.2)
line2, = plt.plot(range(0,201),val_losses_mean)
plt.fill_between(x=range(0,201), y1=val_losses_mean-val_losses_std, y2=val_losses_mean+val_losses_std, alpha=0.2)
plt.xlabel("Epoch", fontsize=12)
plt.xlim(0,total_nr_epochs)
plt.ylabel("Loss", fontsize=12)
plt.legend((line1,line2),("training","validation"))
plt.title('Loss Development with Symbolic Messages', fontsize=15)
plt.grid(True)
#plt.savefig(results_dir+'train_and_val_losses.png')
plt.show()

plt.figure(figsize = (15,7))
line1, = plt.plot(range(0,201),train_accs_mean)
plt.fill_between(x=range(0,201), y1=train_accs_mean-train_losses_std, y2=train_accs_mean+train_accs_std, alpha=0.2)
line2, = plt.plot(range(0,201),val_accs_mean)
plt.fill_between(x=range(0,201), y1=val_accs_mean-val_losses_std, y2=val_accs_mean+val_accs_std, alpha=0.2)
plt.xlabel("Epoch", fontsize=12)
plt.xlim(0,total_nr_epochs)
plt.ylabel("Accuracy", fontsize=12)
#plt.ylim(0.5,1.05)
plt.yticks(np.arange(0.25, 1.05, 0.05))
plt.legend((line1,line2),("training","validation"))
plt.grid(True)
plt.title('Accuracy Development with Symbolic Messages', fontsize=15)
#plt.savefig(results_dir+'train_and_val_accs.png')
plt.show()
```
